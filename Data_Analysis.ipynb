## Processing

La data étant prétraitée, maintenant il faut trouver les corrélations les prédictions toussa tousa (cell 2).

variables accessibles dans le dataframe df : `df['date']`, `df['calls']`, etc. (ou `df.date`, `df.calls`, etc.)


```python
#nb d'appels sur une journée type
plt.plot(df['calls'].groupby(df['date'].dt.hour).sum())
```

```python
#Sur la semaine
# mardi - mercredi - jeudi - vendredi - samedi - dimanche - lundi
plt.plot(df.calls.groupby(df.date.dt.day%7).mean())

#trop simpliste
```

```python
#affiche le profil d'une semaine type
# mardi - mercredi - jeudi - vendredi - samedi - dimanche - lundi
data = (df['calls'].groupby([df.date.dt.day%7,df.date.dt.hour]).sum())
plt.figure()
data.plot(title='Profil semaine en nombre d appels', kind='line')
```

```python
data = (df['duration'].groupby([df.date.dt.day%7,df.date.dt.hour]).mean())
plt.figure()
data.plot(title='Profil semaine en nombre en durée', kind='line')
```
